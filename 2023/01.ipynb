Can you combine first and last into a function?

##### Reply<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

Sure! Here's a function that finds both the first and last digits in a string:

```python
def get_calibration(line):
    first = last = None
    for char in line:
        if char.isdigit():
            if first is None:
                first = char
            last = char
    return first + last
```

This loops through once, setting `first` only on the first digit found, and updating `last` every time. Try it out with your test strings!

````
def get_digits(line):
    first = last = None
    for char in line:
        if char.isdigit():
            if first is None:
                first = char
            last = char
    return int(first + last)
get_digits(x)
````

````
def replace(text):
    replacements = {"one": "o1e", "two": "t2o", "three": "t3e", "four": "f4r", "five": "f5e", "six": "s6x", "seven": "s7n", "eight": "e8t", "nine": "n9e"}
    for old, new in replacements.items():
        text = text.replace(old, new)
    return text
````
