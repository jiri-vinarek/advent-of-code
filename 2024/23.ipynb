A couple of ideas:

1. **Early stopping**: `enumerate_all_cliques` yields cliques in increasing size order. You could use `itertools.takewhile` to stop once cliques exceed size 3, rather than generating all of them (like that size-4 clique at the end).

2. **Direct triangle finding**: For finding just triangles, you could check each edge and see if the two nodes share a common neighbor â€” this avoids generating size-1 and size-2 cliques.

Which approach interests you more?

````
from itertools import *

def solve(inp):
    es = parse(inp)
    G = nx.Graph()
    G.add_edges_from(es)
    cs = L(takewhile(lambda c: len(c) < 4, nx.enumerate_all_cliques(G)))
    return len(cs.filter(lambda o: len(o) == 3).filter(at_least_one_with_t))

solve(samp)
````

Find the largest clique in the graph (the biggest group where every computer is connected to every other). Then output the node names sorted alphabetically, joined by commas.

For the sample, the answer is `co,de,ka,ta`.

````
def solve(inp):
    es = parse(inp)
    G = nx.Graph()
    G.add_edges_from(es)
    cs = L(nx.enumerate_all_cliques(G))
    c = cs[-1]
    return ','.join(sorted(c))

solve(samp)
````
