This word search allows words to be horizontal, vertical, diagonal, written backwards, or even overlapping other words. It's a little unusual, though, as you don't merely need to find one instance of XMAS - you need to find all of them. Here are a few ways XMAS might appear, where irrelevant characters have been replaced with .:

````
rows = L(enumerate(ys))
max_pad_num = len(rows) - 1
padded = rows.map(lambda t: "-"*t[0] + t[1] + "+"*(max_pad_num - t[0]))
padded
````

````
def pad_left(i, s, max_pad_num):
    return "-"*i + s + "+"*(max_pad_num - i)

def pad_right(i, s, max_pad_num):
    return "-"*(max_pad_num - i)  + s + "+"*i

def shift(rows, add_padding_left):
    rows = L(enumerate(rows))
    max_pad_num = len(rows) - 1
    padding_fn = pad_left if add_padding_left else pad_right
    return rows.map(lambda t: padding_fn(t[0], t[1], max_pad_num))

print(list(shift(ys, True)))
print(list(shift(ys, False)))
````

````
def get_column(rows, i):
    rows = L(rows)
    lst = rows.map(lambda s: s[i])
    return "".join(lst)
get_column(ys, 1)
````

````
def get_all_columns(rows):
    n = len(rows[0])
    return L(range(n)).map(lambda i: get_column(rows, i))
get_all_columns(ys)
````

````
def get_all_diags(rows, is_left_diag):
    shifted_rows = shift(rows, is_left_diag)
    return get_all_columns(shifted_rows)
print(get_all_diags(ys, True))
print(get_all_diags(ys, False))
````

````
def get_all_directions(rows):
    return rows + get_all_columns(rows) + get_all_diags(rows, True) + get_all_diags(rows, False)
get_all_directions(ys)
````

How can I count all occurances of a substring in a string?

You can use the `.count()` method on a string! It returns the number of non-overlapping occurrences of a substring.

For example:

```python
"XMASXMAS".count("XMAS")  # Returns 2
```

In your case, you'd want to count both "XMAS" and "SAMX" (the reverse) in each of your directional strings. 

Does that help you move forward with counting all the occurrences across all directions?

````
def count_xmas(rows):
    xs = get_all_directions(rows)
    return L(xs).map(lambda s: s.count("XMAS") + s.count("SAMX")).sum()
count_xmas(xs)
````
