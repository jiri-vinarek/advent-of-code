````
from itertools import takewhile, count

def get_antinode_pos2(a1, a2, rows, cols):
    y1, x1 = a1
    y2, x2 = a2
    dy, dx = y2-y1, x2-x1
    dir1 = L(takewhile(lambda pos: is_in_grid(pos, rows, cols), ((y1 + i*dy, x1 + i*dx) for i in count())))
    dir2 = L(takewhile(lambda pos: is_in_grid(pos, rows, cols), ((y1 - i*dy, x1 - i*dx) for i in count())))
    return dir1 + dir2
get_antinode_pos2((0, 0), (1, 3), 12, 12)
````
