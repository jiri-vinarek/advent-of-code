````
def to_ints(lines):
  return [list(map(int, x.split())) for x in lines.splitlines()]
xs = to_ints(samp)
xs
````

````
x = xs[1]
x
````

````
def diffs(x):
    return list(map(lambda p: p[1] - p[0], list(zip(x, x[1:]))))
diffs(x)
````

````
for x in xs:
    print(diffs(x))
````

````
def is_acceptable(lst):
    return all([x >= 1 and x <= 3 for x in lst]) or all([x <= -1 and x >= -3 for x in lst])
is_acceptable(diffs(xs[0]))
````

````
for x in xs:
    print(is_acceptable(diffs(x)))
````

You can combine them like this:

```python
def is_acceptable(x):
    diffs = [b - a for a, b in zip(x, x[1:])]
    return all(1 <= d <= 3 for d in diffs) or all(-3 <= d <= -1 for d in diffs)
```

This calculates the differences inline and checks the conditions in one function. Want to test it?

````
def is_acceptable(report):
    diffs = [b - a for a, b in zip(report, report[1:])]
    return all(1 <= d <= 3 for d in diffs) or all(-3 <= d <= -1 for d in diffs)
[is_acceptable(x) for x in xs]
````

````
inp_ints = to_ints(inp)
sum([is_acceptable(x) for x in inp_ints])
````

````
def expand(x):
    return [x[0:i] + x[i+1:] for i in range(len(x))]
expand(x)
````

````
[is_acceptable(y) for y in expand(x)]
````

````
sum([not is_acceptable(y) for y in expand(x)]) < 2
````

````
def is_acceptable_with_toleration(report):
    return is_acceptable(report) or sum([not is_acceptable(y) for y in expand(report)]) <= 1
[is_acceptable_with_toleration(x) for x in xs]
````

````
def is_acceptable_with_toleration(report):
    return is_acceptable(report) or any([is_acceptable(y) for y in expand(report)])
[is_acceptable_with_toleration(x) for x in xs]
````

````
sum([is_acceptable_with_toleration(x) for x in inp_ints])
````
