For a directed acyclic graph (DAG), you can use **dynamic programming** to count paths without enumerating them:

- `dp[node]` = number of paths from that node to `out`
- Base case: `dp['out'] = 1`
- For other nodes: `dp[node] = sum(dp[neighbor] for neighbor in successors)`

If you process nodes in reverse topological order, this runs in O(V+E) time â€” much faster than enumeration!

For the constraint (must visit both `dac` and `fft`), you could extend the DP to track *which* required nodes have been visited along the path.
