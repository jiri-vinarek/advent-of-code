````
def parse_c(c):
    if c == '.': return 0
    if c == '@': return 1
    return None
parse_c('@'), parse_c('.')
````

````
def parse_row(x):
    return L(list(x)).map(parse_c)
parse_row(x)
````

````
def parse(s):
    xs = s.splitlines()
    return L(xs).map(parse_row)
x = parse(samp)
x
````

How can I in numpy easily get 8 items surrounding a given position in a 2D array?

Great question! In NumPy, you can use slicing to extract the 8 surrounding neighbors of a position. Here's the pattern:

For a position at `(i, j)`, the 8 neighbors are in a 3x3 grid centered on that position. You can get them with:

```python
neighbors = gp[i-1:i+2, j-1:j+2]
```

This gives you a 3x3 array that includes the center position plus the 8 surrounding cells.

If you only want the 8 neighbors (excluding the center), you could either flatten and skip the middle element, or access specific positions. For counting neighbors (which seems relevant to your problem), you might do:

```python
neighbor_sum = gp[i-1:i+2, j-1:j+2].sum() - gp[i, j]
```

This sums all 9 cells and subtracts the center one.

How can I exclude the middle element?

##### Reply<!-- SOLVEIT_SEPARATOR_7f3a9b2c -->

There are a few ways! Here are two common approaches:

1. **Flatten and skip the middle index:**
   ```python
   neighbors = gp[i-1:i+2, j-1:j+2].flatten()
   neighbors = np.delete(neighbors, 4)  # index 4 is the center in a flattened 3x3
   ```

2. **Use a boolean mask:**
   ```python
   mask = np.ones((3, 3), dtype=bool)
   mask[1, 1] = False
   neighbors = gp[i-1:i+2, j-1:j+2][mask]
   ```

Which approach seems clearer to you, or would you like to see another way?

````
def _neighbors(g, i, j):
    mask = np.ones((3, 3), dtype=bool)
    mask[1, 1] = False
    return g[i-1:i+2, j-1:j+2][mask]
_neighbors(gp, 2, 1)
````

````
cnt = 0
for i in range(1, gp.shape[0]-1):
    for j in range(1, gp.shape[1]-1):
        if gp[i, j] == 1 and _neighbors(gp, i, j).sum() < 4:
            cnt += 1
cnt
````

````
def count_rolls(gp):
    ps = []
    for i in range(1, gp.shape[0]-1):
        for j in range(1, gp.shape[1]-1):
            if gp[i, j] == 1 and _neighbors(gp, i, j).sum() < 4:
                ps.append((i, j))
    return ps
len(count_rolls(gp))
````

````
def solve(inp):
    x = parse(inp)
    g = np.array(x)
    gp = np.pad(g, 1)
    return len(count_rolls(gp))
solve(samp)
````

````
def solve(inp):
    x = parse(inp)
    g = np.array(x)
    gp = np.pad(g, 1)
    gp_orig = gp.copy()
    cnt = 0
    while True:
        ps = count_rolls(gp)
        if not ps: break
        for o in ps: gp[o] = 0
    return int(gp_orig.sum() - gp.sum())
solve(samp)
````
